```python
px.bar(data,x='Date_YMD',y='Daily Confirmed',title='Covid-cases Daily')
```

```python
px.bar(data,x='Date_YMD',y='Daily Deceased',title='Covid-cases Deceased')
```

```python
px.area(data,x='Date_YMD',y='Daily Deceased')
```

```python
cases = data['Daily Confirmed'].sum()
deceased = data['Daily Deceased'].sum()
labels = ['confirmed','Deceased']
values = [cases,deceased]
px.pie(data, values=values, names=labels, title='Daily confirmed v sDaily Death', hole = 0.3)
```

```python
deathrate =(data['Daily Deceased'].sum()/data['Daily Confirmed'].sum() )* 100
print(deathrate)
```

```python
import autots
from autots import AutoTS
model = AutoTS(forecast_length=30,frequency='infer',ensemble='simple')
model = model.fit(data, date_col='Date_YMD', value_col='Daily Deceased',id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
